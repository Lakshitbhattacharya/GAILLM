```python
# 3.0 Read data
data = pd.read_csv('/gdrive/MyDrive/deeplearning/deeplearning/concrete.csv')
data.head()
```

```python
# 3.1
data.shape   # (1030, 9)
print("\n=====\n")
data.dtypes
```

```python
# 3.2 Get target and predictors:
y = data.pop("CompressiveStrength")
X = data
```

```python
# 4.0 Build model:

model = tf.keras.Sequential()
```

```python
# 4.0.1

model.add(layers.Input(shape = (8,) ))
```

```python
# 4.0.2
model.add(layers.Dense(32, activation = 'relu'))
model.add(layers.Dense(32, activation='relu'))
```

```python
# 4.0.4

model.add(layers.Dense(1, activation = 'relu'))
```

```python
# 4.1 Compile:

model.compile( loss = "mae")
```

```python
# 4.2 Train:

model.fit(X,y, epochs = 300 )
```

```python
# 5.0 Predict
pred = model.predict(X)
pred
```
